##### <span style="color:#FFD600">Modeling  

<span style="color:#FFE5B4">(train-test split; training the model with .fit(); generating predictions with .predict() for the training and testing set; evaluating the predictions using metrics): select at least 3 models to try and compare the results.

```python
# Import regression models and evaluation metrics
from sklearn.linear_model import LinearRegression           # Linear regression model
from sklearn.metrics import mean_squared_error, r2_score     # Metrics for model performance
from sklearn.ensemble import AdaBoostRegressor              # Ensemble boosting regressor
import xgboost as xgb                                       # XGBoost regressor
from sklearn.ensemble import RandomForestRegressor          # Random forest regressor
from sklearn.ensemble import GradientBoostingRegressor      # Gradient boosting regressor
from catboost import CatBoostRegressor                      # Cat boosting regressor
```

<span style="color:#B3FFB3"> All Data

<span style="color:#E6B3FF"> Linear Regression

```python
# Create and train a linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train_data, y_train_data)  # fit model to training data
```

```python
# Make predictions using the trained linear regression model
y_test_pred_data = linear_model.predict(X_test_data)    # predictions on test set
y_train_pred_data = linear_model.predict(X_train_data)  # predictions on train set
```

```python
# Evaluate model performance using RÂ² score
print("R2 Linear Regression Train:", r2_score(y_train_data, y_train_pred_data))  # training set
print("R2 Linear Regression Test:", r2_score(y_test_data, y_test_pred_data))    # test set
```

<span style="color:#E6B3FF"> ADAboosting

```python
# Create and train an AdaBoost regression model
abc = AdaBoostRegressor(n_estimators=100, learning_rate=0.05, random_state=42)
model = abc.fit(X_train_data, y_train_data)  # fit model to training data
```

```python
# Make predictions with the trained AdaBoost model
y_test_pred_data = model.predict(X_test_data)    # predictions on test set
y_train_pred_data = model.predict(X_train_data)  # predictions on train set
```

```python
# Evaluate AdaBoost model performance using RÂ² score
print("R2 ADAregressor Train:", r2_score(y_train_data, y_train_pred_data))  # train set
print("R2 ADAregressor Test:", r2_score(y_test_data, y_test_pred_data))    # test set
```

<span style="color:#E6B3FF"> XGboosting

```python
# Create and train an XGBoost regression model
xgb_reg = xgb.XGBRegressor()
xgb_reg.fit(X_train_data, y_train_data)  # fit model to training data
```

```python
# Make predictions with the trained XGBoost model
y_test_pred_data = xgb_reg.predict(X_test_data)    # test set predictions
y_train_pred_data = xgb_reg.predict(X_train_data)  # train set predictions
```

```python
# Evaluate XGBoost model performance using RÂ² score
print("R2 XGboost Train:", r2_score(y_train_data, y_train_pred_data))  # train set
print("R2 XGboost Test:", r2_score(y_test_data, y_test_pred_data))    # test set
```

<span style="color:#E6B3FF"> Random Forest

```python
# Create and train a Random Forest regression model
rf_regressor = RandomForestRegressor(n_estimators=120, random_state=42)
rf_regressor.fit(X_train_data, y_train_data)  # fit model to training data
```

```python
# Predict with Random Forest and calculate MSE
y_test_pred_data = rf_regressor.predict(X_test_data)       # test set predictions
mse_test = mean_squared_error(y_test_data, y_test_pred_data)  # test set MSE

y_train_pred_data = rf_regressor.predict(X_train_data)     # train set predictions
mse_train = mean_squared_error(y_train_data, y_train_pred_data)  # train set MSE
```

```python
# Display training error and performance for Random Forest
print(f"Mean Squared Error: {mse_train:.2f}")                        # training MSE
print("R2 Random Forest Train:", r2_score(y_train_data, y_train_pred_data))  # training RÂ²
```

```python
# Display test error and performance for Random Forest
print(f"Mean Squared Error: {mse_test:.2f}")                     # test set MSE
print("R2 Random Forest Test:", r2_score(y_test_data, y_test_pred_data))  # test set RÂ²
```

<span style="color:#E6B3FF"> Gradient Boosting

```python
# Create and train a Gradient Boosting regression model
gbr = GradientBoostingRegressor(
    loss='absolute_error',  # use MAE as loss function
    learning_rate=0.1,
    n_estimators=300,
    max_depth=1,
    max_features=5,
    random_state=42
)
gbr.fit(X_train_data, y_train_data)  # fit model to training data
```

```python
# Make predictions with the trained Gradient Boosting model
y_test_pred_drop = gbr.predict(X_test_data)     # test set predictions
y_train_pred_drop = gbr.predict(X_train_data)   # train set predictions
```

```python
# Compute Root Mean Squared Error (RMSE) for train and test sets
train_rmse = mean_squared_error(y_train_data, y_train_pred_data) ** 0.5  # training RMSE
test_rmse  = mean_squared_error(y_test_data, y_test_pred_data) ** 0.5   # test RMSE
```

```python
# Evaluate Gradient Boosting model performance using RÂ² score
print("R2 Gradient Train:", r2_score(y_train_data, y_train_pred_data))  # train set
print("R2 Gradient Test:", r2_score(y_test_data, y_test_pred_data))    # test set
```

```python
# Initialize CatBoost Regressor with no training logs (verbose=0) and fixed random seed for reproducibility
cat = CatBoostRegressor(verbose=0, random_state=42)

# Train the model on the training set
cat.fit(X_train_data, y_train_data)
```

```python
# Predicciones en train y test con CatBoost
y_test_pred_data = cat.predict(X_test_data)
y_train_pred_data = cat.predict(X_train_data)
```

```python
# RMSE on training data
train_rmse = mean_squared_error(y_train_data, y_train_pred_data) ** 0.5  

# RMSE on test data
test_rmse  = mean_squared_error(y_test_data, y_test_pred_data) ** 0.5
```

```python
# Print train RMSE
print(f"Mean Squared Train Error: {train_rmse:.2f}")

# Print test RMSE
print(f"Mean Squared Test Error: {test_rmse:.2f}")
```
